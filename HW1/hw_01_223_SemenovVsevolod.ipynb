```python
import numpy as np
import pandas as pd
from scipy import stats
import random as rnd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# your code here ( ⚆ ω ⚆) - моё лицо в попытках понять ТЗ

Kolvo = 10**4
bustype = [5,10]

np.random.seed(seed=1)
s = 0
# а-тый пункт, согрешу, ибо 4 часа раздумий ни к чему не привели...

def workday(s):
  x = list()
  while s < 720:
    l = int(rnd.choice(bustype))
    s += l
    x = x + [l]
    if s == 715:
      l = 5
      s += l
      x = x + [l]

  busses = len(x)
  return busses


arr = np.array([workday(s) for i in range(Kolvo)])

plt.hist(arr, bins=70)
plt.title('Распределение количества автобусов')
plt.ylabel('Количество эксперементов')
plt.xlabel('Количество автобусов в день')

# пункт "Б" написан ниже, после рассуждений над пунктом "А"
```

```python
#my code here... (0_0) Point "Б"

np.random.seed(seed=1)

def passengers(s):
  x = list()
  while s < 720:
    l = int(rnd.choice(bustype))
    s += l
    x = x + [l]
    if s == 715:
      l = 5
      s += l
      x = x + [l]

  busses = x
  return busses # число пассажиров будет равно времени ожидания до прихода автобуса, предполагается, что автобус забирает всех пассажиров мгновенно, первый автобус не учитываем, ибо кто знает, сколько их за ночь пришло**

Timon = np.array([rnd.choice(passengers(s)) for i in range(Kolvo)])

plt.hist(Timon, bins=10)
plt.title('Распределение пассажиров в автобусе с Тимоном')
plt.ylabel('Количество эксперементов')
plt.xlabel('Количество пассажиров')

print("МатОжидание: ", np.mean(Timon))
print("Дисперсия: ", np.var(Timon))
```

В данном случае количество пассажиров равно интервалам между отправлениями, соответственно они могут принимать лишь два значения - 5 и 10. Поскольку эти события равновероятны и независимы, очевидно, что матожидание будет стремиться к среднему арифметическому 5 и 10 (7.5)по мере увеличения выборки. Аналогично для дисперсии, что будет стремиться к 6.25.

```python
# (ง'̀-'́)ง - point "Г|Б"

np.random.seed(seed=1)

def passengersG(s):
  x = list()
  ex = stats.expon(scale=10).rvs(1000)
  while s < 720.0:
    l = np.random.choice(ex)
    s += l
    x = x + [l]

  busses = x
  return busses

TimonG = np.array([rnd.choice(passengersG(s)) for i in range(Kolvo)])

plt.hist(TimonG, bins=100)
plt.title('Распределение пассажиров в автобусах с Тимоном')
plt.ylabel('Количество эксперементов')
plt.xlabel('Количество пассажиров')

print("МатОжидание: ", np.mean(TimonG))
print("Дисперсия: ", np.var(TimonG))
```

В данном случае стоит сказать, что значения выборки независимы и экспоненциально распределенны, потому и гистограммама __напоминает экспоненциальное распределение__, центральный момент присутствует, мы можем говорить о том, что у нас хоть и присутствуют "хвосты", они не такие страшные. Последнее, о чём стоит сказать, поскольку случайные величины у нас непрерывны, а люди приходят раз в минуту, я учитываю только целочисленную часть этих величин, ибо люди по частям не катаются (если только в самых трешовых сценариях)

### Задача 3 (Парадокс Берксона)

Предположим, что результаты ЕГЭ школьников по русскому и математике независимы и хорошо аппроксимируются
нормальным распределением с ожиданием 60 и стандартным отклонением 10.

УШЭ (Урюпинская Школа Экономики) ныне престижна и забирает себе всех школьников набравших более $n$ баллов в сумме по русскому и математике.

Для каждого $n$ от $80$ до $160$ с шагом $5$ случайно создайте $10^4$ школьников и поделите их на прошедших и не прошедших в УШЭ.


__а) [5 баллов]__ Постройте график числа прошедших в УШЭ в зависимости от $n$.

__б) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди прошедших в УШЭ.

__в) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди не прошедших в УШЭ.

```python
# your code here  (ﾉ´ з `)ノ   _-` ┌( ಠ_ಠ)┘    Point (A)
Bachelors = []
Non_Bachelors = []
BachelorsM = []
Non_BachelorsM = []
BachelorsR = []
Non_BachelorsR = []

np.random.seed(seed=15)

MathScore = stats.norm(loc=60, scale=10).rvs(Kolvo)
RusScore = stats.norm(loc=60, scale=10).rvs(Kolvo)
Scores = RusScore + MathScore

Points = np.arange(80, 165, 5)
#Points = Points.reshape(( Points.shape[0],1))

for i in Points:

    luckers = (Scores >= i)
    Bachelors = Bachelors + [Scores[luckers].shape[0]]
    BachelorsM = BachelorsM + [MathScore[luckers]]
    BachelorsR = BachelorsR + [RusScore[luckers]]

    losers = (Scores < i)
    Non_Bachelors = Non_Bachelors + [Scores[losers].shape[0]]
    Non_BachelorsM = Non_BachelorsM + [MathScore[losers]]
    Non_BachelorsR = Non_BachelorsR + [RusScore[losers]]


plt.plot(Points, Bachelors)
plt.title('Распределение поступивших в зависимости от проходных баллов')
plt.ylabel('Количество поступивших')
plt.xlabel('Проходной балл (n)')
```

В данном случае мы видим явную обратную зависимость между количеством поступивших и проходными баллами (Что весьма логично), корреляция между этими величинами скорее всего будет стремиться к "-1"
<br><br>
p.s. подсчитанная мною равнялась -0,97

```python
# ¯\_(ツ)_/¯ Point (Б)

CorrB = []

for j in range(len(Points)):
    corr = stats.pearsonr(BachelorsM[j], BachelorsR[j])
    CorrB = CorrB + [corr[0]]

plt.plot(Points, CorrB)
plt.title('Корреляция между полученными и проходными баллами у поступивших')
plt.ylabel('Значение корреляции')
plt.xlabel('Проходной балл (n)')
```

В данном случае мы можем видеть, что из независимых событий, баллы по русскому и математике стали зависимыми, так как ввели третье событие, с которым они оба связаны, причём, как мы выяснили ранее зависимость обратная. <br>
Она же постепенно начинает проявляться между нашими событиями с увеличением проходных баллов в УШЭ, и чем они больше, тем сильнее заметна отрицательная корреляция

```python
#   ╰(▔∀▔)╯  Point (В)

CorrV = []

for k in range(len(Points)):
    corr = stats.pearsonr(Non_BachelorsM[k], Non_BachelorsR[k])
    CorrV = CorrV + [corr[0]]

plt.plot(Points, CorrV)
plt.title('Корреляция между полученными и проходными баллами у непоступивших')
plt.ylabel('Значение корреляции')
plt.xlabel('Проходной балл (n)')
```

Поскольку непоступлние в УШЭ является обратным событием для заданного в предыдущем пункте, то и корреляция показывает ровно обратные результаты, чем выше баллы, необходимые для поступления, тем она больше
